
Grab relevant config

```python
import sys
import os
from dotenv import load_dotenv

load_dotenv('/kaggle/input/arcprize-2024-common-code/arc_2024/.env')
sys.path.append('/kaggle/input/arcprize-2024-common-code/')

# Now we can set envoriment variables
INPUT_DATA_PATH = os.getenv('INPUT_DATA_PATH')
OUTPUT_DATA_PATH = os.getenv('OUTPUT_DATA_PATH')
TEMP_DATA_PATH = os.getenv('TEMP_DATA_PATH')
TEST_FILE_NAME = os.getenv('TEST_FILE_NAME')

print(INPUT_DATA_PATH)
print(OUTPUT_DATA_PATH)
print(TEMP_DATA_PATH)
print(TEST_FILE_NAME)
```

We run the runner. All the code can be found in the dataset. In general the runner does the following for each task:
1. Uses an 'Interpreter' to see the tasks in 'Shapes'. The hope is that this mimics how humans perceive ARC tasks
2. Uses Induction Logic Programming using FOIL as the core algorithm to find the output grid sizes
3. Using the same process as above but with an extended 'bias' (range of predicates) we aim to solve the ARC tasks

```python
from arc_2024.runner import run

run(
    INPUT_DATA_PATH,
    OUTPUT_DATA_PATH,
    TEMP_DATA_PATH,
    TEST_FILE_NAME,
    split_tasks=True,
    verify_solutions=False,
    submit_empty_solutions=False,
    max_run_time_for_solutions=25200,  # 11.5 hours 41400
    skip_training_data=True)
```
