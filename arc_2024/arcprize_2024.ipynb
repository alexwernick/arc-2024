
Here we inject the config. This is a bit messy as we have to deal with differences between local and Kaggle Setups

```python
import sys
import os
from dotenv import load_dotenv

load_dotenv()
ENVIRONMENT = os.getenv('ENVIRONMENT')

if ENVIRONMENT != 'Local':
    # if not local we assumme we are in a kaggle environment
    load_dotenv('/kaggle/input/arcprize-2024-common-code/.env')
    sys.path.append('/kaggle/input/arcprize-2024-common-code/')

# Now we can set envoriment variables
INPUT_DATA_PATH = os.getenv('INPUT_DATA_PATH')
OUTPUT_DATA_PATH = os.getenv('OUTPUT_DATA_PATH')
TEMP_DATA_PATH = os.getenv('TEMP_DATA_PATH')
TEST_FILE_NAME = os.getenv('TEST_FILE_NAME')
```

The below code can simply be used to explore the input data in the Kaggle environment

```python
import os

if ENVIRONMENT == 'Kaggle':
    for dirname, _, filenames in os.walk('/kaggle/input'):
        for filename in filenames:
            print(os.path.join(dirname, filename))
```

When in Kaggle environment we take the large input files and split in to one file per puzzle

```python
import json
from pathlib import Path
import sys

sys.path.append('/kaggle/input/arcprize-2024-common-code/')

if ENVIRONMENT == 'Local':
    from arc_2024.data_management.data_manager import DataManager
else:
    # if not local we assumme we are in a kaggle environment
    from data_management.data_manager import DataManager
    
data_manager = DataManager(INPUT_DATA_PATH, OUTPUT_DATA_PATH, TEMP_DATA_PATH)

# Split the tasks into individual files
data_manager.split_tasks_to_individual_files(TEST_FILE_NAME)

for unsolved_task in data_manager.get_unsolved_tasks():
    with open(Path(TEMP_DATA_PATH) / f"{unsolved_task}.json", 'r') as file:
        data = json.load(file)
    # Count the number of elements
    element_count = len(data['test'])
    solution = []
    for _ in range(element_count):
        solution.append({"attempt_1": [[0, 0], [0, 0]], "attempt_2": [[0, 0], [0, 0]]})
        
    data_manager.save_individual_solution(solution, unsolved_task)

data_manager.create_solution_file("submission.json")
```
